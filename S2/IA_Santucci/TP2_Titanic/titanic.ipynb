```python
def parse_model_1(train_file):
    train = pd.read_csv(train_file)
    train = train.set_index('PassengerId')
    X = train[['Fare', 'SibSp', 'Parch']]
    y = train['Survived']
    return X, y

X, y = parse_model_1('train.csv')
X.head()
```

2.2.2

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

#2.2.1 on prefere la validation croisÃ©e car elle permet de mieux estimer la performance du modÃ¨le sur des donnÃ©es non vues et avec un echantillon de petite taille
def compute_score(classifier, X, y):
    scores = cross_val_score(classifier, X, y, cv=5)
    return scores.mean()

classifier = LogisticRegression()
compute_score(classifier, X, y)
```

```python
import matplotlib.pyplot as plt

def plot_hist(variable, title):
    plt.hist([dead[variable].dropna(), survived[variable].dropna()], label=['Survived', 'Dead'])
    plt.xlabel(variable)
    plt.ylabel('Number of passengers')
    plt.title(title)
    plt.legend(loc='upper right')
    plt.show()

plot_hist('Pclass','Number of passengers by class')
```

```python
def plot_lr_coefs(X, lr):
    coefs = lr.coef_[0]
    feature_names = X.columns
    plt.bar(feature_names, coefs)
    plt.xlabel('Features')
    plt.ylabel('Coefficient Values')
    plt.title('Logistic Regression Coefficients')
    plt.xticks(rotation=90)
    plt.show()

classifier.fit(X_2, y_2)
plot_lr_coefs(X_2, classifier)

#Question 3.4.4 On constate que les classes 1 et 3 ont un impact plus important que les autres 
#de plus on peut considÃ©rer que le poid de fare est nÃ©gligeable
```

```python
import pylab as pl

def classifier_importance(X, clf):
    importances = clf.feature_importances_
    indices = np.argsort(importances)[::-1]
    pl.title("Feature importances")
    for tree in clf.estimators_:
        pl.plot(range(X.shape[1]), tree.feature_importances_[indices], "r")
    pl.plot(range(X.shape[1]), importances[indices], "b")
    pl.show()
    for f in range(X.shape[1]):
        print("%d. feature: %s (%f)" % (f+1, X.columns[indices[f]], importances[indices[f]]))
```
