```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

iris = pd.read_csv('iris.csv', sep=',')
iris["variety"].value_counts()

#ne pas apprendre et faire les tests sur les mêmes données
```

```python
iris.shape
```

```python
iris.isna().sum() 
#vérification si aucune données manquantes
```

```python
x = iris.drop(columns='variety')
y = iris['variety']
Xtrain, Xtest, Ytrain,Ytest = train_test_split(x,y, test_size=0.33, random_state=0)
```

```python
model = SVC(C=100)
model.fit(Xtrain, Ytrain)
ypred = model.predict(Xtest)
model.score(Xtest, Ytest)
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(Xtrain, Ytrain)
ypred = model.predict(Xtest)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(Ytest, ypred), annot=True, xticklabels=model.classes_, yticklabels=model.classes_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
score = []
for k in range(1, 30):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(Xtrain, Ytrain)
    score.append(model.score(Xtest, Ytest))
```

```python
plt.plot(range(1, 30), score)
```

```python
model = KNeighborsClassifier(n_neighbors=6)
model.fit(Xtrain, Ytrain)
ypred = model.predict(Xtest)
sns.heatmap(confusion_matrix(Ytest, ypred), annot=True, xticklabels=model.classes_, yticklabels=model.classes_)
```
