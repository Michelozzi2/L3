```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
capitales = pd.read_csv('Capitales.csv', sep=';', index_col='Villes')
capitales = capitales.iloc[:, range(0, 12)]
capitales
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=2)
capitales_pca = model.fit_transform(capitales)
capitales_pca = pd.DataFrame(capitales_pca,index = capitales.index)
capitales_pca
```

```python
ratio = pd.DataFrame(model.explained_variance_ratio_)
ratio
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

capitales_cah = linkage(capitales, method='ward', metric='euclidean')
plt.title("CAH")

dendrogram(capitales_cah, labels= capitales.index, orientation='right', color_threshold=0)
plt.show
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(distance_threshold=None,n_clusters=5)
capitales_cah = model.fit(capitales)
```

```python
capitales_cah.labels_
```

```python
CAH = pd.DataFrame({'label': capitales_cah.labels_}, index=capitales.index)
CAH
```

```python
capitales_pca = capitales_pca.iloc[:,:2]
```

```python
model =AgglomerativeClustering(distance_threshold=None,n_clusters=6)
model.fit(capitales_pca)
```

```python
CAH["label_acp"] = model.labels_
CAH
```

```python
from sklearn.cluster import KMeans
inertie = []
for k in range(1,20) :
    KMean = KMeans(n_clusters=k)
    KMean.fit(capitales)
    inertie.append(KMean.inertia_)
plt.plot(range(1,20), inertie)

```
